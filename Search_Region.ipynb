```python
keywords = pd.read_csv('keywords.csv')
```

```python
def geo_data(keyword):
    geo_dict = {
        'Центр': ['москва', 'moscow', 'тула', 'tula', 'ярославль', 'yaroslavl'],
        'Северо-Запад': ['петербург', 'petersburg', 'saint petersburg', 'st petersburg',
                         'псков', 'pskov', 'мурманск', 'murmansk'],
        'Дальний Восток': ['владивосток', 'vladivostok', 'сахалин', 'sakhalin',
                           'хабаровск', 'khabarovsk']
    }
    key = keyword.lower().strip()

    for region, cities in geo_dict.items():
        for city in cities:
            if city in key:
                return region

    return 'undefined'

keywords['region'] = keywords['keyword'].apply(geo_data)

keywords.sort_values('region', ascending=False).head()
```

```python
print(geo_data('moscow'))
```

```python
print(geo_data('суперджоб хабаровск'))
```
